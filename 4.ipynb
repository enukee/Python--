```python
import numpy as np
import csv
```

```python
# открытие файла data.csv как матрицы numpy
my_data = np.genfromtxt("data.csv", delimiter=',')
print(my_data)
```

```python
# количество измерений
print(my_data.ndim)
```

```python
# размерность матрицы
print(my_data.shape)
```

```python
# количество элементов
print(my_data.size)
```

```python
# тип элементов
print(my_data.dtype)
```

```python
# первый элемент первой строки
print(my_data[0,0])
```

```python
# последняя строка
print(my_data[-1:,:])
```

```python
# каждый 5 элемент последней строки
print(my_data[-1:, ::5])
```

```python
# вывод 6 первых строк
print(my_data[:6, :])
```

```python
# среди первых шести строк вывод кажой пятой
print(my_data[:6:5,:])
```

```python
# обнуление 1 строки
my_data[0] = 0
print(my_data)
```

```python
# вывод суммы элемнтов
print(my_data.sum())
```

```python
# вывод среднего значения
print(my_data.mean())
```

```python
# вывод медианы
print(np.median(my_data))
```

```python
# вывод стандартного отклонения
print(np.std(my_data))
```

```python
# нормализация
print((my_data- np.min(my_data)) / (np.max(my_data) - np.min(my_data)))
```

```python
# стандартизация
np.set_printoptions(precision=8, suppress=True, formatter={'all': lambda x: f'{x:0.8f}'})
print((my_data- np.mean(my_data)) / (np.std(my_data)))
```
